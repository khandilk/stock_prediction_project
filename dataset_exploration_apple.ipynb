```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download real Apple stock data
data = yf.download("AAPL", start="2010-01-01", end="2023-11-13")
close_prices = data['Close']

# Plot the real closing prices
plt.figure(figsize=(12, 5))
plt.plot(close_prices, label='Apple Close Price')
plt.title("Apple (AAPL) Closing Prices (2010–2023)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("apple_real_close_prices.png")  # Save it for your slides
plt.show()
```
